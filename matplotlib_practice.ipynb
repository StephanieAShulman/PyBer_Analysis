```python
# Chain pyplot to matplotlib
#  Variation of - from matplotlib import plyplot
#  Use standard alias - plt
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Import array data - months, US dollars/mo
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

### Chart Creation: MATLAB Method

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

### Chart Creation: Object-Oriented Interface

```python
# fig, ax = plt.subplots()
# ax.plot(x_axis, y_axis)

# Create the plot with ax.plt()
fig, ax = plt.subplots()
# plt.subplots() returns tuple with figure/axies objects
# fig, ax = subplots() unpacks the tuple into fig, ax
ax.plot(x_axis, y_axis)
# ax.plot(x,y) plots data from x,y axes
# fig, ax = plt.subplots() can also be written as
#   #create plot with ax.plot()
#   fig = plt.figure()
#   ax = fig.add_subplot()
```

```python
# Create a plot with ax.plot()
fig = plt.figure()
ax = fig.add_subplot()
# Empty plot prints until line below ...
ax.plot(x_axis, y_axis)
```

```python
# Via OOI
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

### Change the Major Ticks

```python
# When the dollars were original plotted, major ticks on the x-axis were spaced every $10.
#  But this makes it difficult to estimate average fares.
#  Incremental $5 could help

import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis() # get current axes & invert
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
# Can also use
# ax.xasix.set_ticks()
# If y-axis had numerical values, could change how often every major y-tick
#  occurs with yticks() for MATLAB and set_yticks() for OOI.
```

### Minor Ticks

```python
# Can be added witout numerical values between major ticks on the x or y-axis.
# set_minor_locator fxn with xaxis class (OOI)
# x.xaxis.set_minor_locator()
# Add MultipleLocator() and pass how often minor ticks wanted
#  Note: Have to import matplotlib.ticker fxn
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
